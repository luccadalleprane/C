```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

N = 1000
x = np.linspace(-1, 1, N)
y = 2*x + 3
y = y + 0.10*np.random.randn(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
print(x.shape)
print(y.shape)

(1000,1)
(1000,1)
```

```python
xtrain, xvalid, ytrain, yvalid = train_test_split(x, y, test_size = 0.3)
print(xtrain.shape)
print(xvalid.shape)

(700,1)
(300,1)
```

```python
plt.scatter(xtrain, ytrain)
plt.scatter(xvalid, yvalid)
plt.show()
```

```python
neuronio = Sequential()
neuronio.add(Dense(1))
neuronio.compile(optimizer = 'sgd' , loss = 'mse')
```

```python
r = neuronio.fit(xtrain , ytrain , epochs = 100, batch_size = 32)
```

```python
plt.plot(r.history['loss'])
plt.show()
```

```python
neuronio.evaluate(xvalid, yvalid)
```

```python
ypred = neuronio.predict(xvalid)
```

```python
plt.scatter(xtrain, ytrain)
plt.scatter(xvalid, ypred)
plt.show()
```
